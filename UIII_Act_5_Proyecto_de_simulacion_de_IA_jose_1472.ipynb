```python
# Bienvenida e interacción inicial
print("--- Bienvenido al Salón de Belleza 'Estilo Inteligente' ---")
presupuesto = float(input("Para comenzar, ingrese su presupuesto disponible: "))

# Selección de servicio principal
print("Servicios: 1. Corte de cabello ($30), 2. Tinte ($70), 3. Tratamiento ($50)")
opcion = int(input("¿Qué servicio desea realizarse? (Ingrese el número): "))

# Procesamiento de la opción de Corte (Uso de operadores relacionales y lógicos)
if opcion == 1 and presupuesto >= 30.0:
    print("Has elegido: Corte de cabello.")
    largo = input("¿Desea un corte 'Corto' o 'Largo'? ")

    # Decisión basada en el estilo
    if largo == "Corto" or largo == "corto":
        print("La IA sugiere un estilo 'Pixie' o 'Bob' para su rostro.")
    else:
        print("La IA sugiere un corte en capas para mantener el volumen.")

    # Cálculo aritmético del cambio
    balance = presupuesto - 30.0
    print(f"Servicio completado. Su saldo restante es: ${balance}")

# Procesamiento de la opción de Tinte
elif opcion == 2 and presupuesto >= 70.0:
    print("Has elegido: Tinte profesional.")
    color = input("¿Prefiere tonos 'Cálidos' o 'Fríos'? ")

    if color == "Cálidos" or color == "cálidos":
        print("IA: Le recomendamos un tono cobrizo o dorado.")
    elif color == "Fríos" or color == "fríos":
        print("IA: Le recomendamos un tono cenizo o platino.")
    else:
        print("IA: Opción no reconocida, procederemos con un tono neutro.")

    balance = presupuesto - 70.0
    print(f"Proceso de color terminado. Saldo restante: ${balance}")

# Procesamiento de la opción de Tratamiento
elif opcion == 3 and presupuesto >= 50.0:
    print("Has elegido: Tratamiento Capilar.")
    daño = int(input("Del 1 al 10, ¿qué tan dañado siente su cabello? "))

    # Operador relacional para determinar intensidad del tratamiento
    if daño >= 7:
        print("IA: Aplicaremos una máscara de reconstrucción profunda.")
    else:
        print("IA: Aplicaremos un tratamiento de hidratación ligera.")

    balance = presupuesto - 50.0
    print(f"Tratamiento aplicado. Saldo restante: ${balance}")

# Manejo de casos de error o presupuesto insuficiente (Operador relacional y lógico)
elif opcion > 3 or opcion < 1:
    print("Error: El servicio seleccionado no existe en nuestro catálogo.")
else:
    print("Lo sentimos, su presupuesto es insuficiente para el servicio seleccionado.")
    falta = 0.0
    if opcion == 1:
        falta = 30.0 - presupuesto
    elif opcion == 2:
        falta = 70.0 - presupuesto
    elif opcion == 3:
        falta = 50.0 - presupuesto
    print(f"Le faltan ${falta} para completar el pago.")

print("--- Gracias por usar Estilo Inteligente ---")
```
